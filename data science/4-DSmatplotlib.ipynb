Matplotlib is a Python library for drawing charts. It is a must learn for beginners in data visualisation. plt is the conventional alias for matplotlib.pyplot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

The standard code for drawing a Matplotlib chart is <code>plt.<i>chart_type</i>(<i>parameters</i>)</code>. The parameters are typically some Python iterables, the number of parameters depending on the chart type. For example, for a histogram there is only one parameter, an iterable of numbers. In the following example, I generate 100 random numbers in a Numpy array from a normal distribution, then plot a histogram to see if they indeed follow a normal distribution:

```python
a = np.random.normal(10,5,100)
plt.hist(a)
```

To add labels to a chart, add them after the code drawing the chart. For example to add title of the chart and names of x- and y-axis: 

```python
plt.hist(a)
plt.title("Random numbers from a normal distribution")
plt.xlabel("Value of random number")
plt.ylabel("Frequency")
```

The full list of customizations you can make to the chart can be found at https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html?highlight=hist#matplotlib.pyplot.hist, as can the full documentation to the hist function. Read if you are interested.

Another common chart type that accepts one parameter is the pie chart. The parameter is an iterable of numbers, each number related to the size of a sector:

```python
a = [3.9,4.41,7,14.2]
plt.pie(a)
```

To add names to the sectors, set a keyword argument, or optional parameter <code>label</code>in plt.pie equal to another iterable with the same length, consisting of names that correspond to the numbers at the same position:

```python
a = [3.9,4.4,7.1,14.2]
names = ['John','Paul','Ringo','George']
plt.pie(a,labels=names)
```

Most customizations for other charts are usually specified using keyword arguments as well.

For a chart type that accepts two parameters, the first is usually the x-axis and the second the y-axis, each represented by an iterable of the same length. For example,

```python
x = [0,1,2,3]
y = [10,0.2,-4.3,5]
```

are two iterables of the same length, so you can draw a chart based on these two iterables. For example, the <code>scatter</code> function in Matplotlib will draw a scatter graph of four points, the x and y of which are elements of the x and y lists at the same position, ie. (0,10), (1,3.2), (2,-4.3) and (3,5):

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

And as alluded before, customizations are specified using keyword arguments, for example if I want to change the color of the markers:

```python
plt.scatter(x,y,c='r')
```

The <code>plot</code> function creates a line chart of the four points:

```python
plt.plot(x,y,linestyle='--')
```

Then, to for example draw a graph of any mathematical function, we can simply define a Numpy array of values of x which is the range in which we want to draw the graph, with interval as small as you like, and then express y as a mathematical function of x, then use plot. For example for a graph of y = x cos(x):

```python
x = np.arange(-7,7,0.05)
y = x*np.cos(x)
plt.plot(x,y)
```

For a vertical bar chart, the x-axis are an iterable of categories, and the y-axis another iterable with the same length of numbers. We can for example turn the pie chart above into a bar chart:

```python
a = [3.9,4.4,7.1,14.2]
names = ['John','Paul','Ringo','George']
plt.bar(names,a,color='orange')
```

Matplotlib can also work with data in a Pandas dataframe. For example, we can select a column in a dataframe to be the x-axis and another the y-axis, and draw a scatter chart:

```python
df
```

```python
import pandas as pd
df = pd.read_csv("iris.csv")
a = df["SepalLength"]
b = df["SepalWidth"]
plt.scatter(a,b,marker='x',c='purple')
```

To draw two (or more) graphs in the same chart, simply run the two commands sequentially.

```python
t = np.arange(0.0, 5.0, 0.1)
plt.plot(t, np.exp(-t)*np.cos(2*np.pi*t), color='blue', marker='o')
plt.plot(t, np.cos(2*np.pi*t), color='orange', linestyle='--')
```

To draw two graphs but in separate areas, it is better to first use the subplots function. The first parameter in the function is the number of rows of chart areas, and the second the number of columns. The function returns two results, conventionally named fig and axs. Note the syntax on plotting the graphs relating to the axs object.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0, 5.0, 0.1)
fig, axs = plt.subplots(2,1)

axs[0].plot(t, np.exp(-t)*np.cos(2*np.pi*t), color='blue', marker='o')

axs[1].plot(t, np.cos(2*np.pi*t), color='orange', linestyle='--')
plt.show()
```

It is also possible to draw 3-dimensional plots. The syntax is more complicated, so I shall not explain here. Read it up yourself why the syntax is this way if you are interested.

```python
df = pd.read_csv("iris.csv")
a = df["sepal.length"]
b = df["sepal.width"]
c = df["petal.length"]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(a,b,c)
```
