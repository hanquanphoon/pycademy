<p> Seaborn is another Python library used for chart drawing. It is a wrapped around Matplotlib, but has significant advantages of simplifying a lot of command by more abstraction and default settings for many commonly used charts and combinations of charts. </p>
<p> The conventional alias used when importing Seaborn is sns. The sns.set() command changes the default chart style to that of a more modern-looking Seaborn style. </p>

```python
import seaborn as sns
import numpy as np
sns.set()
```

Seaborn can work with Numpy array in the way similar to Matplotlib:

```python
x = np.arange(0,7,0.1)
y = np.sin(x)
sns.lineplot(x,y)
```

But the biggest advantage of Seaborn is that it works conveniently with Pandas. In the parenthesis of a Seaborn chart command, simply write data = df will specify the dataframe used in the command is df. The x-axis, y-axis, and color distinction are similarly easily and intuitively specified.

```python
sns.scatterplot(x="sepal_length",y="sepal_width",hue="species",data=df)
```

Another commonly drawn chart type is a scatter plot along with a line of best fit of the data points. In Matplotlib the two charts have to be manually plotted out. In Seaborn, there is a simple chart type called lmplot that does it automatically for you:

```python
g = sns.lmplot(x="sepal_length", y="sepal_width", hue="species", data=df)
```

Another important chart type used in data science provided by Seaborn is pairplot. It can show all pairwise scatter charts of all numerical columns in a dataframe, giving the data scientist an excellent overall glance of relations between any two attributes, along with the distributions of each attribute.

```python
sns.pairplot(hue="species",data=df)
```

And for simple charts like a bar chart, Seaborn intelligently guesses what you want even if your command is vague, for example here it guesses you want the average of some data, and also gives you the error bar:

```python
sns.barplot(x="species",y="sepal_width",data=df)
```

And many more. You should read the documentation for Seaborn for more things you can do with Seaborn!
